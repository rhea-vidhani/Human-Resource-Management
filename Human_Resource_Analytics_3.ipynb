```python
# Create variable SpecialProjectsCount
def bucket_projects(x):
    if x <= 1:
        return "Low (0–1)"
    elif 2 <= x <= 3:
        return "Medium (2–3)"
    else:
        return "High (4+)"

df['SpecialProjectsBucket'] = df['SpecialProjectsCount'].apply(bucket_projects)
```

```python
#Create variable Lateness_Group
def group_lateness(x):
    if x == 0:
        return "Perfect Attendance (0)"
    elif x in [1, 2]:
        return "Slightly Late (1-2)"
    elif x in [3, 4]:
        return "Moderately Late (3-4)"
    elif x in [5, 6]:
        return "Highly Late (5-6)"
    else:
        return "Other"

df['Lateness_Group'] = df['DaysLateLast30'].apply(group_lateness)
```

```python
#Create variable group_absences
def group_absences(x):
    if x == 0:
        return "No Absences (0)"
    elif 1 <= x <= 5:
        return "Low Absences (1-5)"
    elif 6 <= x <= 10:
        return "Moderate Absences (6-10)"
    elif 11 <= x <= 15:
        return "High Absences (11-15)"
    elif 16 <= x <= 20:
        return "Critical Absences (16-20)"
    else:
        return "Above 20 Absences"

df['Absence_Group'] = df['Absences'].apply(group_absences)
```
