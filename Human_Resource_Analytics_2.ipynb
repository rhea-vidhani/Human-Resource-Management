```python
#Create variable EngagementGroup
def engagement_group(score):
    if score <= 2.0:
        return "Very Low Engagement"
    elif score <= 3.0:
        return "Low Engagement"
    elif score <= 4.0:
        return "Moderate Engagement"
    else:
        return "High Engagement"

df['EngagementGroup'] = df['EngagementSurvey'].apply(engagement_group)
```
